```r
# Coloured by trimmedf PHRED score
## Make colours
col = "trimmed.PHRED"
colours_vir = viridis_pal(option = "D")(length(unique(df[[col]])))
colours_hue = c("#7f64b9", "#70ad5d", "#b94a74", "#bb7638")

# Make plot
plot_phred = ggplot(df, aes(x=chi.square.statistics, y=ANOVA.F.statistics, group=trimmed.PHRED)) +
  geom_point(aes(color=trimmed.PHRED), size=5, shape=20) +
  guides(colour = guide_legend(override.aes = list(size=15))) +
  scale_color_manual(values=colours_vir) + # Manual colours
  xlab("Chi-Square statistics") + 
  ylab("ANOVA F-statistics") +
  labs(colour='trimmed PHRED score')
  #theme_classic()
  #theme_bw()
  #geom_text(label=df$pipeline) # Labels
  
# Plot plot in preview
plot_phred

# Save plot in workdir
jpeg("plot_phred.jpg", width = 1200, height = 800)
plot_phred
dev.off()
```
